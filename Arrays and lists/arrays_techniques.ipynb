```python

from typing import List

def remove_element(arr: List[int], ele: int) -> int:
    k = 0
    for i in range(len(arr)):
        if arr[i]!= ele:
            arr[k], arr[i] = arr[i], arr[k]
            k += 1
    return k


arr = [0, 1, 3, 0, 2, 2, 4, 2]
ele = 2
print(remove_element(arr, ele))
```

```python
# this method utilizes the 3 pointer method. one is i, i-1 and idx


def removeDuplicates(arr):
    n = len(arr)
    if n <= 1:
        return n

    # Start from the second element
    idx = 1  
    
    for i in range(1, n):
        if arr[i] != arr[i - 1]:
            arr[idx] = arr[i]
            idx += 1

    return idx
```
